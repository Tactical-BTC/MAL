#### 핵 보유 개수에따라 csv파일 만들기

```python
def MN_det(df_list, number):
    Alphabet = ['A','B','C','D','E','F','G','H','I']
    for i in range(len(df_list)):
        df = df_list[i].query('MITO_count == {} and Nucleus_count == {}'.format(number,number))
        name = 'N{}M{}'.format(number,number) + '_' + Alphabet[i] + '.csv'
        df.to_csv(name, header=True, index=False)
```

```python
def MN(df_list, number):
    Alphabet = ['A','B','C','D','E','F','G','H','I']
    df_total = []
    for i in range(len(df_list)):
        df = df_list[i].query('MITO_count == {} and Nucleus_count == {}'.format(number,number))
        df_total.append(df)
        
    return df_total
```

### Total = N1M1 + N2M2로 정의

```python
N1M1 = MN(Total, 1)
N2M2 = MN(Total, 2)
```

```python
N1M1_total = reduce(lambda x,y: pd.concat([x,y],axis=0), N1M1)
N1M1_total.reset_index().iloc[:,1:]
```

```python
N2M2_total = reduce(lambda x,y: pd.concat([x,y],axis=0), N2M2)
N2M2_total.reset_index().iloc[:,1:]
```

```python
Total = pd.concat([N1M1_total,N2M2_total], axis=0).reset_index().iloc[:,1:]
Total
```
