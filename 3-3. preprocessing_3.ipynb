### RandomUnderSampling, random_state=0, sampling_strategy='auto'

```python
X = Total_df.iloc[:,6:-1]
y = Total_df.iloc[:,[-1]]
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=0, sampling_strategy='auto')
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

```python
X_resampled
```

```python
Total_trim = pd.concat([X_resampled,y_resampled],axis=1)
```

```python
Total_trimed = pd.merge(Total_trim,Total_df, how = 'left')
```

```python
Total_trimed
```

```python
col1 = Total_trimed.iloc[:, -5:-2]
col1
```

```python
col2 = Total_trimed.iloc[:, :-7]
col2
```

```python
col3 = Total_trimed.iloc[:, [-7]]
col3
```

```python
Final_Total = pd.concat([col1, col2, col3],axis=1)
```
