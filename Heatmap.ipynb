```python
Final = [pd.read_csv(i, index_col=0) for i in Analysis_data]
```

```python
Final = [Final[i].iloc[:,3:] for i in range(len(Final))]
```

```python
def total_df(df, scaler = StandardScaler):
    target = df.loc[:,['labels']]
    df_num = df.loc[:,:'RBC_Texture SER Dark 0 px']
    scaler = scaler()
    data = scaler.fit_transform(df_num.values)
    df_norm = pd.DataFrame(data, columns = df_num.columns, index = df.index)
    df_total = pd.concat([df_norm, target],axis=1)
    Total_median = df_total.groupby('labels').median()
    
    return Total_median
```

```python
Total_median = list(map(total_df, Final))
```

```python
Time = ['A (06 - 12 h)', 'B (11 - 17 h)','C (16 - 21 h)','D (21 - 27 h)','E (28 - 34 h)','F (32 - 38 h)','G (38 - 42 h)','H (42 - 48 h)']
Time_bold = ['<b>'+Time[i]+'</b>' for i in range(0,len(Time))]
Time_bold
```

```python
def Heatmap(num):
        Time = ['A (06 - 12 h)', 'B (11 - 17 h)','C (16 - 21 h)','D (21 - 27 h)','E (28 - 34 h)','F (32 - 38 h)','G (38 - 42 h)','H (42 - 48 h)']
        Time_bold = ['<b>'+Time[i]+'</b>' for i in range(0,len(Time))]
        fig = go.Figure(data=go.Heatmap(
                z= Total_median[num],
                x = Total_median[num].columns,
                y = Time_bold,
                colorscale='YlGnBu'))
        fig.update_xaxes(side="top")
        fig.update_yaxes(autorange="reversed")
        fig.update_layout(height=600, width=2100,template='plotly_white')
        fig.show()
```
